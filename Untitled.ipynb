```python
import pandas as pd
import numpy as np
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
import re
```

```python
data_string = '''Observations     706         706         706         706    
R2              0.132       0.131       0.128       0.007   
AdjustedR2     0.121       0.123       0.121       0.002  '''
df = pd.read_csv(io.StringIO(data_string) ,
    delimiter='\s+', header=None, index_col=0)
df
```

```python
df=df.transpose() #серия=столбец
df
```

```python
df['m'] = [10,8,6,5]
df
```

```python
df['F_obs'] = df['R2']/(1 - df['R2'])*(df['Observations']-df['m'])/(df['m']-1)
df
```

```python
alpha = 0.05
df['F_cr'] = stats.f.ppf(1 - alpha,df['m']- 1, df['Observations'] - df['m'])
df
```

```python
df['Znachimost'] = df.apply(lambda x : 'znachimo' if x['F_obs'] >x['F_cr'] else 'ne znachimo', axis = 1 )
df
```
